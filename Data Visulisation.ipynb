Hexplot using Seaborn

Heatmaps using Seaborn

```python
corrmat = df2.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8, square=True)
```

Boxen Plot using Seaborn

```python
sns.catplot(x="body_style", y="stroke", data=df2, kind="boxen",height=4, aspect=2.7, hue = "drive_wheels")
```

Ridge Plot using seaborn

```python
# Initialize the FacetGrid object
g = sns.FacetGrid(df2, row="symboling", hue="drive_wheels", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "symboling", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "symboling", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "symboling")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
